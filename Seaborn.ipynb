```python
import seaborn as sns 
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
planets = sns.load_dataset("planets")
```

```python
tips 
```

```python
sns.scatterplot(data=tips,x="tip",y="total_bill",hue="day",size="size",palette="YlGnBu")
```

```python
sns.histplot(tips["tip"],kde=True,bins=15)#data=tips,x="tip", bunu yazmak yerine kÄ±saca tips["tip"]
```

```python
sns.displot(tips["tip"],kde=True,bins=15)
```

```python
sns.barplot(x="sex",y="tip",data=tips,palette="YlGnBu",hue="sex")
```

```python
sns.boxplot(x="day",y="tip",data=tips,hue="sex",palette="YlGnBu")
```

```python
sns.boxplot(x="day",y="total_bill",hue="sex",data=tips,palette="YlGnBu")
```

```python
sns.stripplot(x="day",y="tip",data=tips,hue="sex",palette="YlGnBu",dodge=True)
```

```python
sns.jointplot(x="tip",y="total_bill",data=tips,kind="reg")
```

```python
sns.jointplot(x="tip",y="total_bill",data=tips,kind="kde",fill=True,cmap="YlGnBu")
```

```python
sns.jointplot(x="tip",y="total_bill",data=tips,kind="hex",cmap="YlGnBu")
```

```python
sns.pairplot(titanic.select_dtypes(["number"]),hue="pclass")
```

```python
titanic_corr = titanic.corr(numeric_only=True)
titanic_corr
```

```python
sns.heatmap(titanic_corr,annot=True,cmap="coolwarm")
```
